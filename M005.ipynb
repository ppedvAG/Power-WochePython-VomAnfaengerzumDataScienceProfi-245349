```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # WICHTIG: Unterpaket pyplot importieren
```

```python
data = pd.read_csv("Data/AmazonStock.csv", parse_dates=True, index_col="Date")
```

```python
data
```

### Plotten über Pandas

```python
data.plot()
```

```python
data["Close"].plot()
```

### Plotten über matplotlib

```python
plt.plot(data["Close"])
```

```python
# Plot konfigurieren
plt.figure(figsize=(12, 3))
plt.xlabel("Year")
plt.ylabel("Close Price")
plt.grid(alpha=0.2)
plt.plot(data["Close"], color="Red")

plt.show()  # Hier wird der Konfigurationsprozess beendet

plt.plot(data["Open"] * 2)
```

### Globale vs. Objektorientierte Graphen

Bei globalen Graphen werden alle Befehle über plt. ausgeführt

Bei plt.show() oder am Ende der Zelle werden alle Befehle auf den Plot angewandt

Beim objektorientierten Graph wird eine Variable angelegt, und der Plot wird in diese Variable gespeichert

```python
fig, ((a1_1, a1_2), (a2_1, a2_2)) = plt.subplots(2, 2)

a1_1.bar(np.arange(10), np.random.randint(10, size=(10)))
a1_2.plot(np.random.random(size=100))
a2_1.pie(np.random.randint(5, size=4))
a2_2.hist(np.random.randint(10, size=1000))
```

### Tortendiagramm

```python
pop = pd.read_csv("Data/PopulationDataFertig.csv")
```

```python
plt.pie(pop["WorldShare"][:10])

None  # None um den Rückgabewert auszublenden
```

```python
plt.figure(figsize=(3, 3))
plt.bar(pop["Country"][:10], pop["Pop"][:10])
plt.xticks(rotation=60)

plt.show()

plt.figure(figsize=(3, 3))
plt.barh(pop["Country"][:10], pop["Pop"][:10])

None
```

```python
countries = pop.iloc[:20]
plt.figure(figsize=(12, 4))
plt.bar(countries["Country"], countries["Pop"], color="green", edgecolor="black", linewidth=1)
plt.xticks(rotation=45)
plt.grid(alpha=0.3, linestyle="--")
None
```

### Histogramm

Verteilungsdiagramm

```python
x = np.random.randint(100, size=1000)
```

```python
plt.hist(x, color="green", edgecolor="black", linewidth=1, bins=20)  # bins: Anzahl Balken (Gruppierung von Werten)
plt.xticks(np.arange(20) * 5)
None
```
