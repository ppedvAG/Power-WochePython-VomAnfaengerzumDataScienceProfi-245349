#### Funktion für N.A. definieren

```python
def fixNA(columnName: str, newValue):
    zeilen = data[data[columnName] == "N.A."]  # Finde die fehlerhaften Zeilen
    index = zeilen.index
    data.loc[index, columnName] = newValue
```

```python
fixNA("MedAge", 0)
```

```python
fixNA("UrbanPopPct", 0)
```

Diese Funktion kann auf jede Spalte angewandt werden, da sie keine Änderungen verursacht, wenn die Spalte fehlerfrei ist

```python
for col in data.columns:
    fixNA(col, 0)
```

### Prozentzeichen

```python
data["WorldShare"].replace(" %", "", regex=True)  # regex=True muss hier verwendet werden
```

Mit data["Spaltenname"] = kann eine neue Spalte hinzugefügt/bestehende Spalte überschrieben werden

```python
data["WorldShare"] = data["WorldShare"].replace(" %", "", regex=True)
```

```python
data
```

```python
def fixPct(columnName: str):
    data[columnName] = data[columnName].replace(" %", "", regex=True)
```

```python
for col in data.columns:
    fixPct(col)
```
